```python
# March Mania 2025 / Tutorial [Japanese]
# https://www.kaggle.com/code/takuji/march-mania-2025-tutorial-japanese

import os 
import warnings
import pandas as pd

warnings.simplefilter('ignore')
pd.set_option('display.max_columns', None)
```

```python
# コンペデータの確認
for dirname, _, filenames in os.walk('./data'):
    if dirname != "./data":
        continue
    filenames = sorted(filenames)
    print("コンペデータのCSVの数は", len(filenames))
    cnt = 0
    for filename in filenames:
        cnt += 1
        print("*" * 50)
        try:
            df = pd.read_csv(os.path.join(dirname, filename))
        except:
            df = pd.read_csv(os.path.join(dirname, filename), encoding='cp1252')
        print(f"No.{cnt}", filename, df.shape)
        display(df.head(3))
```
