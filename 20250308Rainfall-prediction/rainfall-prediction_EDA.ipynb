## EDA(探索的データ分析)の目的

* データ理解  
    * データのサンプルの確認 
    * データの形状、型の把握   
    * 統計分析(平均値、中央値、標準偏差、最大値、最小値、)
    * データの可視化(分布、相関、散布図)
* データクリーニング
    * データの修正(欠損値、異常値)
    * 重複データの処理
    * カテゴリ変数の処理
    * データの型変換
    * データのスケーリング
* 特徴量エンジニアリング
    * 新しい特徴量の作成
    * bin化
    * 次元削減
* 仮説を立てる
    * モデル決定
    * 評価手法決定
    * その他
